```
import numpy as np
from math import e
import pandas as pd
from neural_net import neural_network
import matplotlib.pyplot as plt
```

```
df1 = pd.read_csv("sample_data/mnist_train_small.csv",header = None)
```

```
df2 = pd.read_csv("sample_data/mnist_test.csv",header = None)
```

```
y_train = df1[0]
y_test = df2[0]
```

```
x_train = df1.drop(0,axis = 1)
x_train.describe()
```

```
x_test = df2.drop(0,axis= 1)
x_test.describe()
```

```
model = neural_network()
```

```
# 4 layer neural network 
model.fit(x_train, y_train, k = 10, layers = (20,20,10,20), epochs = 1000, lamda = 0.1, alpha = 0.05)
```

```
model.plot_loss()
```

```
# Train Accuracy 
y_pred = model.predict(x_train)
model.accuracy(y_pred,y_train)
```

```
# Test Accuracy
y_pred = model.predict(x_test)
model.accuracy(y_pred,y_test)
```

```
from sklearn.neural_network import MLPClassifier
mlp = MLPClassifier(hidden_layer_sizes = (20,20,10,20), max_iter = 1000)
mlp.fit(x_train, y_train)
```

```
y_pred = mlp.predict(x_test)
np.mean(y_pred == y_test)
```
