```
import numpy as np
import pandas as pd
from dtree import dtree
from collections import Counter
```

```
df1 = pd.read_csv('/content/sample_data/mnist_train_small.csv', header = None)
df1.head()
```

```
df2 = pd.read_csv('/content/sample_data/mnist_test.csv', header = None)
df2.head()
```

```
y_train = df1[0]
y_test = df2[0]
```

```
x_train = df1.drop(0, axis = 1)
x_test = df2.drop(0, axis = 1)
```

```
%%time
model = dtree()
model.fit(x_train, y_train, min_ig = 0.04)
```

```
y_pred = model.predict(x_test)
model.accuracy(y_test,y_pred)
```
