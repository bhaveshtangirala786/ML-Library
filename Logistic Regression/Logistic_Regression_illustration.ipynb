```
import numpy as np
import pandas as pd
from math import e
from math import log
import matplotlib.pyplot as plt
from logistic_reg import logistic_regression
import warnings
warnings.filterwarnings("ignore")
```

```
df1 = pd.read_csv("sample_data/mnist_train_small.csv",header = None)
```

```
df2 = pd.read_csv("sample_data/mnist_test.csv",header = None)
```

```
df1.head()
```

```
y_train = df1[0]
y_test = df2[0]
```

```
x_train = df1.drop(0, axis = 1)
x_train.describe()
```

```
x_test = df2.drop(0, axis = 1)
x_test.describe()
```

```
model = logistic_regression()
```

```
model.train(x_train, y_train, epochs = 1000, k = 10)
```

```
model.plot_loss()
```

```
# Train Accuracy
y_pred = model.predict(x_train).flatten()
model.accuracy(y_pred,y_train)
```

```
# Test Accuracy
y_pred = model.predict(x_test).flatten()
model.accuracy(y_pred,y_test)
```

```
from sklearn.linear_model import LogisticRegression
```

```
model1 = LogisticRegression()
model1.fit(x_train, y_train)
```

```
# Test Accuracy by using sklearn
model1.score(x_test, y_test)
```
